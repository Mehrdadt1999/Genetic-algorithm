```python
import numpy as np
import random
State_space = np.random.randint(0, 8, (30, 8))

random_indices = np.random.choice(State_space.shape[0], size = 24, replace = False)
cr = State_space[random_indices, :]


for i in range(1, 24, 2):
  block_index = random.randint(0,8)
  child1 = np.append(cr[i][0 : block_index], cr[i-1][block_index : 8])
  child2 = np.append(cr[i-1][0:block_index], cr[i][block_index:8])

  child1 = child1.reshape(1, 8)
  child2 = child2.reshape(1, 8)

  State_space = np.append(State_space, child1, axis = 0)
  State_space = np.append(State_space, child2, axis = 0)

random_indices = np.random.choice(State_space.shape[0], size = 16, replace = False)
mu = State_space[random_indices, :]
# print(mu)
for i in range(0, 16):
  rand_index = random.randint(0, 7)
  rand_num = random.randint(0, 9)
  mu[i][rand_index] = rand_num


cost = np.zeros((54,1))
print(cost)
# cost = cost.reshape(54, 1)
# print(cost)
counter = 0
print(State_space)
for i in State_space:
  for j in range(0, 8):
    for k in range(j + 1, 8):
      if(i[j] == i[k]):
        cost[counter] += 1
      elif (i[j] - i[k] == j - k):
        cost[counter] += 1
  counter += 1

State_space = np.append(State_space, cost, axis = 1)



print(cost)
print(cost.shape)
print(State_space)
print()
State_space = State_space[np.argsort(State_space[:, 8])]
print(State_space)
State_space = np.delete(State_space, 8, axis = 1)
print(State_space)
State_space = State_space[0:20]
print(State_space)
```
